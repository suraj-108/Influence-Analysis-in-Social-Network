```python
he = nx.from_pandas_edgelist(h, source = "0", target = "1")

nx.info(he)
```

degree centrality means that people who have been most influential in the group. Here it means the the super heroes who have appeared the most in the marvel comics books. It also signifies the number of times the super heroes appear in the comic books.

```python
DC1=[]
MF = nx.degree_centrality(he)
i=0
counter = 0
for w in sorted(MF, key = MF.get , reverse = True):
    counter = counter + 1
    if counter == 6:
        break
    DC1.append(w)
    i-i+1

DC1
```

eigenvector specifies which super hero knows other super heros the most./ has appeared in other comics / helped the other heroes in that comic

```python
EV1=[]
EV = nx.eigenvector_centrality(he)
counter = 0
for w in sorted(EV, key =EV.get,  reverse = True):
    counter = counter + 1
    if counter == 6:
        break
    EV1.append(w)

EV1
```

```python
BC1=[]
BC = nx.betweenness_centrality(he)
counter = 0
for w in sorted(BC, key =BC.get,  reverse = True):
    counter = counter + 1
    if counter == 6:
        break
    BC1.append(w)

BC1
```

```python
CC1=[]
CC = nx.closeness_centrality(he)
counter = 0
for w in sorted(CC, key =CC.get,  reverse = True):
    counter = counter + 1
    if counter == 6:
        break
    CC1.append(w)

CC1
```

```python
PG1=[]
PG = nx.pagerank(he)
counter = 0
for w in sorted(PG, key =PG.get,  reverse = True):
    counter = counter + 1
    if counter == 6:
        break
    PG1.append(w)

PG1
```

```python
h= h.iloc[:10000,:2]
he = nx.from_pandas_edgelist(h, source = "0", target = "1")
print(nx.info(he))
```

```python
import igraph as ig

#Numbers of Nodes
N = he.number_of_nodes()

#List of Edge
L = he.number_of_edges()

#Graph objects
Edges_name = [e for e in he.edges()] #Edges Names

Edges= nx.convert_node_labels_to_integers(he) #Mapping all Nodes into Numbers
Edges = [e for e in Edges.edges()]

#Graph
G = ig.Graph(Edges, directed = False)

# Geolocalization
layt = G.layout('kk',dim = 3)  #3D Localization

#Given X,y,z Position
Xn=[layt[k][0] for k in range(N)]# x-coordinates of nodes
Yn=[layt[k][1] for k in range(N)]# y-coordinates
Zn=[layt[k][2] for k in range(N)]# z-coordinates

Xe=[]
Ye=[]
Ze=[]

#Grouping Coordinates
for e in Edges:
    Xe+=[layt[e[0]][0],layt[e[1]][0], None]# x-coordinates of edge ends
    Ye+=[layt[e[0]][1],layt[e[1]][1], None]
    Ze+=[layt[e[0]][2],layt[e[1]][2], None]
    
#Nodes Name 
labels = []
group = []

for i in range(len(Edges_name)):
    value = Edges_name[i][0]
    labels.append(value)
    
for i in range(len(Edges)):
    value = Edges[i][0]
    group.append(value)
    
group =[]
group.extend(np.repeat(1,2000))
group.extend(np.repeat(2,2000))
group.extend(np.repeat(3,3000))
group.extend(np.repeat(4,1000))
group.extend(np.repeat(5,2000))

trace1=go.Scatter3d(x=Xe,y=Ye,z=Ze,mode='lines',
                    line=dict(color='rgb(125,125,125)', width=1),hoverinfo='none')

trace2=go.Scatter3d(x=Xn,y=Yn,z=Zn,mode='markers',name='SuperHeroe',
                    marker=dict(symbol='circle',size=4,color=group,colorscale='Viridis',line=dict(color='rgb(50,50,50)', width=0.5)),
                    text=labels,hoverinfo='text')

axis=dict(showbackground=False,showline=False,zeroline=False,
          showgrid=False,showticklabels=False,title='')
```

```python
layout = go.Layout(
         title="Network Of appaerance",
    autosize=True,
         width=1000,
         height=1000,
         showlegend=False,
         scene=dict(
             xaxis=dict(axis),
             yaxis=dict(axis),
             zaxis=dict(axis),
        ),
     margin=dict(
        t=100
    ),
    hovermode='closest',
    annotations=[
           dict(
           showarrow=False,
            xref='paper',
            yref='paper',
            x=0,
            y=0.1,
            xanchor='left',
            yanchor='bottom',
            font=dict(
            size=14
            )
            )
        ],    )

data=[trace1, trace2]
fig=go.Figure(data=data, layout=layout)
fig.show()
```
